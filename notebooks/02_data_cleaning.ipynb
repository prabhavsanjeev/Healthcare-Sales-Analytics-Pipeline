# Healthcare Data Analytics Project
## Step 2: Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
# loading our raw data-
data=pd.read_csv(r"../data/raw/healthcare_data.csv")
```

```python
data.head()
```

```python
# for checking the shape of the data-
data.shape  
```

```python
# for checking the info of the data-
data.info()

# INFO SHOWS - We have 2 columns i.e Insurance cover and Doctor_Assigend where exists 25 missing values
```

```python
# for the statistical analysis of the data-
data.describe()

# SHOWS - It will show us the basic stats of the integer datatype columns like thier count, mean, std., 25% etc...
```

```python
# now we have to check the missing values-
data.isna().sum(axis=0)
```

```python
#  now let us fill the missing values on the column named - Doctor_Assigned -
data.fillna({"Doctor_Assigned":"Unknown"},inplace=True)
```

```python
# to fill the missing values in the Insurance_Cover-
data.fillna({"Insurance_Cover":data["Insurance_Cover"].median()},inplace=True)
```

```python
#  verify the missing values again -
data.isna().sum()
```

```python
# for checing the duplicates -
data.duplicated().sum()
```

```python
# if exists - to remove the dupliactes -
data.drop_duplicates(keep=False,inplace=True)
```

```python
# Lets calculate the cost per day- total cost / days of stay-
data["Cost_per_Day"] = data["Final_Bill"] / data["Length_of_Stay"]
data
```

```python
# create age groups-
data["Age_Group"] = pd.cut(
    data["Age"],
    bins=[0, 18, 40, 60, 100],
    labels=["Child", "Young", "Adult", "Senior"]
)
data
```

```python
# covert the columns that are having dates into date format datatype-
data["Admission_Date"] = pd.to_datetime(data["Admission_Date"])
data["Discharge_Date"] = pd.to_datetime(data["Discharge_Date"])
data
```

```python
# Final check our cleaned dataset-
data.head()
```

```python
# lets save this clean data in our "Processed" file of data-
data.to_csv("../data/processed/cleaned_healthcare_data.csv", index=False)
```
