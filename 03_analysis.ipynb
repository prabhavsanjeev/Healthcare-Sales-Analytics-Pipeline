# Healthcare Data Analytics Project
## Step 3: Data Analysis & Visualization

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# lets laod the data that we have made for analyis i.e. the cleaned version:) - 
data=pd.read_csv("../data/processed/cleaned_healthcare_data.csv")
data
```

```python
# Lets analyse this data on the basis of questions -
# 1. For the total number of patients-
data["Patient_ID"].nunique()  # will tell the number of diffent patients i.e there will be no one repeating
```

```python
# 2. Average treatment cost -
data["Treatment_Cost"].mean()
```

```python
# 3. Which is the most common disease -
data["Disease"].value_counts()  # from this we can check the disaese which is most common will be repeated several times, we have checked it is - COVID -19 repaeted 109 times i.e 109 patients are there suffering from COVID 19
```

```python
# 4. For the gender distribution -
data["Gender"].value_counts()
```

# Let's dig deeper-


```python
# 5. Avg cost per disease -
data.groupby("Disease")["Treatment_Cost"].mean()
```

```python
# 6. Average stay by department-
data.groupby("Hospital_Department")["Length_of_Stay"].mean()
```

```python
# 7. Revenue generated by city-
data.groupby("City")["Final_Bill"].sum()
```

# VISUALIZE THE THINGS THAT WE HAVE DONE SO FAR-


```python
# For the diseases and their numbers-
data["Disease"].value_counts().plot(kind="bar")
plt.title("Disease Distribution")
plt.xlabel("Disease")
plt.ylabel("Count")
plt.show()
```

```python
data["Outcome"].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Patient Outcome Distribution")
plt.show()
```

```python
# for the percentage of patients suffering from different diseases-
data["Disease"].value_counts().plot(kind="pie",autopct="%1.1f%%")
```
