# Healthcare Data Analytics Project
## Step 1: Data Creation
In this notebook, we generate a synthetic healthcare dataset using Pandas and NumPy.

```python
# to create the data- here I will create the rows my myslef without directly fetching any already stored data-
import pandas as pd
import numpy as np
# For reproducibility
np.random.seed(42)

# Number of records
n = 500

# Generate data
data = pd.DataFrame({
    "Patient_ID": np.arange(1000, 1000 + n),
    "Age": np.random.randint(1, 90, n),
    "Gender": np.random.choice(["Male", "Female"], n),
    "City": np.random.choice(["Jaipur", "Delhi", "Mumbai", "Bangalore"], n),
    "Disease": np.random.choice(["Diabetes", "Heart Disease", "Flu", "Cancer", "COVID-19"], n),
    "Doctor_Assigned": np.random.choice(["Dr. Sharma", "Dr. Mehta", "Dr. Khan", "Dr. Singh"], n),
    "Hospital_Department": np.random.choice(["Cardiology", "General", "Oncology", "ICU"], n),
    "Treatment_Cost": np.random.randint(5000, 50000, n),
    "Insurance_Cover": np.random.randint(1000, 30000, n),
    "Payment_Status": np.random.choice(["Paid", "Pending"], n),
    "Outcome": np.random.choice(["Recovered", "Critical", "Deceased"], n)
})

# Create admission dates
data["Admission_Date"] = pd.to_datetime(
    np.random.choice(pd.date_range(start="2023-01-01", end="2023-12-31"), n)
)

# Length of stay
data["Length_of_Stay"] = np.random.randint(1, 15, n)

# Discharge date
data["Discharge_Date"] = data["Admission_Date"] + pd.to_timedelta(data["Length_of_Stay"], unit='D')

# Final Bill Calculation
data["Final_Bill"] = data["Treatment_Cost"] - data["Insurance_Cover"]

# Add some missing values intentionally
for col in ["Doctor_Assigned", "Insurance_Cover"]:
    data.loc[data.sample(frac=0.05).index, col] = np.nan

# Shuffle dataset
data = data.sample(frac=1).reset_index(drop=True)
```

```python
data.to_csv("../data/raw/healthcare_data.csv", index=False)
```
